# Task 2: Exploratory Data Analysis (EDA)

In this section, I will analyze the cleaned dataset to extract meaningful business insights.


```python
import pandas as pd

df = pd.read_csv("output/cleaned_data.csv")

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
total_sales = df['sales'].sum()
total_sales
```

```python
df.groupby('order_year')['sales'].sum()
```

```python
df.groupby('region')['sales'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

df.groupby('order_year')['sales'].sum().plot(kind='bar')
plt.title("Sales by Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.groupby('category')['sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('sub-category')['sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('customer_name')['sales'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('category')['sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
df['shipping_days'].plot(kind='hist', bins=20)
plt.title("Shipping Days Distribution")
plt.show()
```
